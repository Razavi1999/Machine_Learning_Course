# Classification

```python
import scipy.io
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FastICA
from sklearn.manifold import TSNE
from mne.decoding import CSP
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

file_path = '/content/drive/MyDrive/ML_Project/BCICIV_1_mat/BCICIV_calib_ds1a.mat'
data = scipy.io.loadmat(file_path)

eeg_signals = data['cnt']
eeg_markers = data['mrk']
eeg_info = data['nfo']

# 1. Low-pass Filter
def low_pass_filter(data, cutoff=40, fs=250, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = signal.butter(order, normal_cutoff, btype='low', analog=False)
    filtered_data = signal.filtfilt(b, a, data, axis=0)
    return filtered_data

filtered_signals = low_pass_filter(eeg_signals)

# 2. Spatial Filtering

# Common Average Reference (CAR) Filtering
def car_filter(data):
    car_signal = data - np.mean(data, axis=1, keepdims=True)
    return car_signal

car_signals = car_filter(filtered_signals)

# Laplacian Filtering (simplified example)
def laplacian_filter(data):
    laplacian_signal = data - np.mean(data, axis=1, keepdims=True)
    return laplacian_signal

laplacian_signals = laplacian_filter(filtered_signals)

# PCA
pca = PCA(n_components=10)
pca_signals = pca.fit_transform(filtered_signals)

# ICA
ica = FastICA(n_components=10)
ica_signals = ica.fit_transform(filtered_signals)

# 3. t-SNE for Signal Visualization
def plot_tsne(data, labels, title):
    tsne = TSNE(n_components=2, random_state=42)
    tsne_result = tsne.fit_transform(data)
    plt.figure(figsize=(10, 6))
    plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=labels, cmap='viridis', s=2)
    plt.colorbar()
    plt.title(title)
    plt.show()

# Reduce dimensions before t-SNE
pca_for_tsne = PCA(n_components=50)
reduced_signals = pca_for_tsne.fit_transform(filtered_signals)

# Correctly accessing the labels
labels = eeg_markers[0][0][1].flatten()
# plot_tsne(reduced_signals, labels, "t-SNE Visualization of Filtered EEG Signals")

# 4. Feature Extraction with CSP (Common Spatial Patterns)
csp = CSP(n_components=4, reg=None, log=True, norm_trace=False)
labels = np.where(labels == -1, 0, labels)
csp_features = csp.fit_transform(filtered_signals, labels)

# Visualize CSP features
# plot_tsne(csp_features, labels, "t-SNE Visualization of CSP Features")

# 5. Classification
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(csp_features, labels, test_size=0.3, random_state=42)

# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred)
print("Logistic Regression Accuracy:", lr_acc)
print(classification_report(y_test, lr_pred))

# Support Vector Machine
svm = SVC()
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
print("SVM Accuracy:", svm_acc)
print(classification_report(y_test, svm_pred))

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_acc)
print(classification_report(y_test, rf_pred))

# K-Nearest Neighbors
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
print("KNN Accuracy:", knn_acc)
print(classification_report(y_test, knn_pred))

print("Classification complete.")
```
